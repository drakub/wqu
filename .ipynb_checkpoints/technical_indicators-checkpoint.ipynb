### Step 1: Import the necessary packages

```python
import yfinance as yf

import sys
import pandas as pd
import math
import numpy as np

import matplotlib
from matplotlib import pyplot as plt
from matplotlib import style
import seaborn as sns

from datetime import datetime
from dateutil.relativedelta import relativedelta
```

### Step 2: Download Data from Yahoo

```python
commodity_tickers = ['KC=F', 'CT=F','ZC=F', 'ZS=F', 'ZW=F']
indicator_tickers = ['NATR', 'DX', 'SMA', 'ATR', 'AD']
```

```python
start_date = '2011-07-01'
stop_date = '2016-06-30'
```

```python
comm_prices = yf.download(commodity_tickers, start=start_date, end=stop_date)['Adj Close']
```

```python
indi_values = yf.download(indicator_tickers, start=start_date, end=stop_date)['Adj Close']
```

### Step 3: Pre-Processing, Analysis and Visualization

#### a) A peak through the datasets

```python
# visualization through the datasets (head, tail)
comm_prices.head()
```

```python
comm_data_df = comm_prices.copy()
comm_data_df.reset_index('Date', inplace=True)

commodity_tickers = {
    'CT=F' : 'Cotton',
    'KC=F' : 'Coffee',
    'ZC=F' : 'Maize',
    'ZS=F' : 'Soybeans',
    'ZW=F' : 'Wheat'
}

for key, name in commodity_tickers.items():
    comm_data_df.rename(columns = {key:name}, inplace = True)
```

```python
comm_data_df.head()
```

```python
comm_columns = ['Cotton', 'Coffee', 'Maize', 'Soybeans', 'Wheat']
```

```python
min_max_scaled_df = comm_data_df.copy()
for column in comm_data_df:
    num_ = (min_max_scaled_df[column] - min_max_scaled_df[column].min())
    den_ = (min_max_scaled_df[column].max() - min_max_scaled_df[column].min())
    min_max_scaled_df[column] = num_/den_
    
plt.figure(figsize=(12, 6))
plt.plot(min_max_scaled_df)
plt.ylabel('Adjusted Close Prices \n(US Dollars)')
plt.xlabel('Period (Months)')
plt.legend(comm_columns)
plt.show()
```

```python
indi_values.head()
```

```python
indi_data_df = indi_values.copy()
indi_data_df.reset_index('Date', inplace=True)
```

```python
indi_data_df.head()
```

```python
indi_columns = ['AD', 'ATR', 'DX', 'NATR', 'SMA']
```

```python
min_max_scaled_df = indi_data_df.copy()
for column in indi_data_df:
    num_ = (min_max_scaled_df[column] - min_max_scaled_df[column].min())
    den_ = (min_max_scaled_df[column].max() - min_max_scaled_df[column].min())
    min_max_scaled_df[column] = num_/den_
    
plt.figure(figsize=(12, 6))
plt.plot(min_max_scaled_df)
plt.ylabel('Adjusted Close Prices \n(US Dollars)')
plt.xlabel('Period (Months)')
plt.legend(indi_columns)
plt.show()
```

#### c) Data Analysis and Visualization

```python
## Combining the datasets
new_df = comm_data_df.merge(indi_data_df, how='right')
```

```python
new_df.head()
```

```python
new_df_columns = ['Cotton', 'Coffee', 'Maize', 'Soybeans', 'Wheat', 'AD', 'ATR', 'DX', 'NATR', 'SMA']

min_max_scaled_new_df = new_df.copy()
for column in new_df:
    num_ = (min_max_scaled_new_df[column] - min_max_scaled_new_df[column].min())
    den_ = (min_max_scaled_new_df[column].max() - min_max_scaled_new_df[column].min())
    min_max_scaled_new_df[column] = num_/den_

plt.figure(figsize=(12, 6))
plt.plot(min_max_scaled_new_df)
plt.ylabel('Adjusted Close Prices \n(US Dollars)')
plt.xlabel('Period (Months)')
plt.legend(new_df_columns)
plt.show()
```

```python
new_df_corr = new_df.corr()
print("Correlation Matrix is : ")
print(new_df_corr)
```

```python
fig = plt.figure(figsize=(15, 9))
sns.heatmap(new_df_corr, annot = True)
plt.show()
```
